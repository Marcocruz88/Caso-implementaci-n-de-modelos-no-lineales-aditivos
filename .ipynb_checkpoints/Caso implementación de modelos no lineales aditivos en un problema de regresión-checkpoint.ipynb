### 2. Estimacin de modelos lineales  (25 puntos)
Encuentren el modelo lineal que mejor predice la variable "Sales" entre Lasso, PCA, PLS. Para esto, estimen los modelos usando la muestra de entrenamiento y evalen el poder predictivo en la muestra de test. Describan su procedimiento y analicen los resultados.

```python
catcols = df.select_dtypes(exclude=['int64', 'float64']).columns
df_dummies = pd.get_dummies(df, columns=catcols, dtype=int, drop_first=True)
df_dummies
```

```python
XTotal = df_dummies.drop(columns=['Sales'], errors='ignore').to_numpy()
YTotal = df_dummies['Sales'].to_numpy()
```

```python
#Divisin de train y test con 33% en test
XTrain, XTest, yTrain, yTest = train_test_split(XTotal, YTotal, test_size=0.33, random_state=0)
```

```python
scaler = StandardScaler()
XTrain_scale = scaler.fit_transform(XTrain)
XTest_scale  = scaler.transform(XTest)
```

Lasso

```python
#### Utilicen este espacio para escribir los cdigos del procedimiento del punto 2 ####
alphasCalibrar = np.logspace(-10, 2, 200)

modeloLASSO = LassoCV(alphas=alphasCalibrar).fit(XTrain_scale, yTrain)
R2LASSO=modeloLASSO.score(XTrain_scale, yTrain)
coeficientesLASSO=modeloLASSO.coef_
prediccionLASSO=modeloLASSO.predict(XTest_scale)
MSELASSO=np.average(np.square(prediccionLASSO-yTest))
print('MSE regresin por LASSO:')
print(MSELASSO)
```

PCA con metodo forward

```python
n_features = XTrain.shape[1]
kf = KFold(n_splits=5, shuffle=True, random_state=42)

seleccionados = []
disponibles = list(range(n_features))
mejor_mse = np.inf
historial_mse = []

while disponibles:
    mejor_comp = None
    mejor_mse_temp = np.inf
    
    for comp in disponibles:
        prueba = seleccionados + [comp]
        mses_folds = []
        
        # Cross-validation 
        for train_idx, val_idx in kf.split(XTrain):
            X_tr, X_val = XTrain[train_idx], XTrain[val_idx]
            y_tr, y_val = yTrain[train_idx], yTrain[val_idx]
            
            # Scaler dentro del fold
            scaler = StandardScaler()
            X_tr_s = scaler.fit_transform(X_tr)
            X_val_s = scaler.transform(X_val)
            
            # PCA full dentro del fold
            pca = PCA(n_components=n_features)
            Z_tr = pca.fit_transform(X_tr_s)
            Z_val = pca.transform(X_val_s)
            
            # Seleccionar PCs especficos
            Z_tr_sub = Z_tr[:, prueba]
            Z_val_sub = Z_val[:, prueba]
            
            # Regresin
            lr = LinearRegression()
            lr.fit(Z_tr_sub, y_tr)
            y_pred = lr.predict(Z_val_sub)
            mses_folds.append(mean_squared_error(y_val, y_pred))
        
        mse_cv = np.mean(mses_folds)
        if mse_cv < mejor_mse_temp:
            mejor_mse_temp = mse_cv
            mejor_comp = comp
    
    # Criterio de parada
    if mejor_mse_temp < mejor_mse - 1e-6:
        seleccionados.append(mejor_comp)
        disponibles.remove(mejor_comp)
        mejor_mse = mejor_mse_temp
        historial_mse.append(mejor_mse)
        print(f"Paso {len(seleccionados)}: Agregado PC{mejor_comp + 1}, MSE_CV = {mejor_mse:.4f}")
    else:
        print("\nNo hay mejora al agregar ms componentes. Deteniendo.")
        break

print("RESULTADO:")
print("Componentes seleccionados:", [f"PC{c+1}" for c in seleccionados])
print("Nmero de componentes:", len(seleccionados))
```

```python
pca_final = PCA(n_components=n_features)
Z_train = pca_final.fit_transform(XTrain_scale)
Z_test = pca_final.transform(XTest_scale)
lr_final = LinearRegression().fit(Z_train[:, seleccionados], yTrain)

y_pred_train = lr_final.predict(Z_train[:, seleccionados])
y_pred_test = lr_final.predict(Z_test[:, seleccionados])

mse_test_pca = mean_squared_error(yTest, y_pred_test)
r2_test_pca = r2_score(yTest, y_pred_test)


print('MSE regresin por PCA con metodo forwward:')
print(mse_test_pca)
```

PLS

```python
pls = PLSRegression(n_components=XTrain.shape[1])
pls.fit(XTrain, yTrain)
prediccionPLS=pls.predict(XTest)
prediccionPLS=np.hstack(prediccionPLS)
MSEPLS=np.average(np.square(prediccionPLS-yTest))
print('MSE regresin por partial least squares:')
print(MSEPLS)
```

Ridge

```python
# Definir rango de alphas similares a Lasso
alphas_ridge = np.logspace(-10, 2, 200)
# Entrenar Ridge con Cross-Validation
modeloRidge = RidgeCV(alphas=alphas_ridge, cv=5).fit(XTrain_scale, yTrain)
# Predicciones y mtricas
coeficientesRidge = modeloRidge.coef_
prediccionRidge = modeloRidge.predict(XTest_scale)
MSERidge = mean_squared_error(yTest, prediccionRidge)
print('MSE regresin por Ridge:')
print(MSERidge)
print(f"Mejor alpha Ridge: {modeloRidge.alpha_}")
```

```python
# Comparacin de los MSE de los modelos estimados
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
ax.set_title("Comparacin de Modelos - MSE (Menor es mejor)")


ejeX = ['LASSO', 'PCA Forward', 'PLS', 'Ridge']
ejeY = [MSELASSO, mse_test_pca, MSEPLS, MSERidge]

# Colores para diferenciar
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

bars = ax.bar(ejeX, ejeY, color=colors)

# Etiquetas de valor sobre las barras
def addlabels(x, y, plotP):
    for i in range(len(x)):
        plotP.text(i, y[i] + 0.01, f"{y[i]:.4f}", ha='center', fontweight='bold')

addlabels(ejeX, ejeY, plt)

# Lnea de referencia del mejor modelo hasta ahora (Lasso probablemente)
min_mse = min(ejeY)
plt.axhline(y=min_mse, color='gray', linestyle='--', alpha=0.7)

plt.ylabel('Mean Squared Error (MSE)')
plt.show()

# Conclusin automtica impresa
best_model_idx = np.argmin(ejeY)
print(f"\n El mejor modelo es {ejeX[best_model_idx]} con un MSE de {ejeY[best_model_idx]:.4f}")
```

#### Anlisis y Conclusin 2

**Procedimiento:**
* Inicialmente se determinaron cules eran las variables categricas y se separaron en variables independientes, con el fin de posteriormente aplicar one-hot encoding mediante variables dummy y as conservar toda la informacin, segn lo observado en el EDA.
* Se realiz una separacin entre conjuntos de entrenamiento y prueba, utilizando un 33% del dataset para prueba.
* Se utiliz StandardScaler para escalar XTrain y XTest, debido a que, como se observ en el EDA, las variables presentaban escalas muy diferentes y los modelos podan verse afectados por esta situacin.

**LASSO:** Se implement un modelo de regresin LASSO, utilizando un intervalo de 200 valores de alpha para calibrar el hiperparmetro de regularizacin. Para el entrenamiento se emplearon las variables predictoras previamente estandarizadas, con el fin de evitar sesgos derivados de diferencias en escala entre variables.

**PCA:** Se aplic PCA sobre los datos de entrenamiento estandarizados para obtener todas las componentes principales posibles. Adicionalmente, para reducir el sesgo asociado a seleccionar el nmero de componentes nicamente con base en la varianza explicada, se utiliz el mtodo de seleccin secuencial forward, que realiza un proceso iterativo iniciando con un conjunto vaco de componentes seleccionados. En cada iteracin se evala la incorporacin de un nuevo componente al modelo. Se utiliz adems validacin cruzada de 5 folds para estimar el MSE en cada configuracin y seleccionar el conjunto de componentes que minimiza el MSE con validacin cruzada. Finalmente, este proceso se repite hasta que no haya una mejora significativa segn un criterio de parada (mejora < 1e-6).

**PLS:** Se aplic PLS sobre los datos de entrenamiento; en este caso no fue necesario utilizar datos previamente estandarizados, ya que el modelo realiza el escalamiento internamente. Por lo tanto, se entren utilizando los datos originales.

**Resultados:** 
* MSE lasso de 1.2329
* MSE PCA + forward de 1.2664
* MSE PLS de 1.2664

De acuerdo con los resultados obtenidos para este dataset, el modelo LASSO presenta un rendimiento superior frente a PCA y PLS, mostrando una mejora de 0.0335 unidades en el MSE, mientras que PCA y PLS presentan prcticamente el mismo error. Estos resultados pueden estar derivados de la metodologa utilizada por cada modelo. Mientras que PCA y PLS buscan reducir la dimensionalidad del espacio de variables para simplificar el modelo, LASSO aplica una penalizacin que permite seleccionar automticamente las variables ms relevantes. En este sentido, como se observ en el EDA, este dataset ya se encuentra bien estructurado y las variables originales contienen informacin significativa para la prediccin, por lo tanto, aplicar una regularizacin mediante penalizacin puede resultar ms efectivo que reducir dimensionalidad y potencialmente perder informacin valiosa para el modelo.

#### Modelo MARS

```python
# ============================================================
# 3.a) Modelo MARS - Multivariate Adaptive Regression Splines
# ============================================================
# Enfoque: Se generan funciones base tipo hinge h(x-t) y h(t-x)
# en cada variable, y se usa LassoCV para seleccionar las
# funciones relevantes (equivalente al pruning de MARS).
#
# Referencia: Friedman (1991) - MARS
# Similar al uso de mars() del paquete mda en R.
# ============================================================

feature_names = ['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
                 'Age', 'Education', 'ShelveLoc_Good', 'ShelveLoc_Medium',
                 'Urban_Yes', 'US_Yes']

# --- Paso 1: Generar funciones base hinge (candidatas) ---
n_knots = 15  # Knots por variable
hinge_info = []  # (direccion, feature_idx, knot)

for j in range(XTrain_scale.shape[1]):
    knots = np.percentile(XTrain_scale[:, j], np.linspace(10, 90, n_knots))
    for t in knots:
        hinge_info.append(('pos', j, t))   # h(x_j - t)+
        hinge_info.append(('neg', j, t))   # h(t - x_j)+

def build_hinge_matrix(X, hinge_info):
    """Construir matriz de funciones base hinge"""
    cols = []
    for direction, j, t in hinge_info:
        if direction == 'pos':
            cols.append(np.maximum(0, X[:, j] - t))
        else:
            cols.append(np.maximum(0, t - X[:, j]))
    return np.column_stack(cols)

B_train = build_hinge_matrix(XTrain_scale, hinge_info)
B_test = build_hinge_matrix(XTest_scale, hinge_info)

print(f"Funciones base hinge generadas: {B_train.shape[1]}")
print(f"  ({n_knots} knots x {XTrain_scale.shape[1]} variables x 2 direcciones)")

# --- Paso 2: Seleccion de funciones base con LassoCV ---
alphas_mars = np.logspace(-4, 1, 100)
mars_model = LassoCV(alphas=alphas_mars, cv=5, max_iter=10000)
mars_model.fit(B_train, yTrain)

# --- Paso 3: Resultados ---
pred_mars_test = mars_model.predict(B_test)
mse_mars = mean_squared_error(yTest, pred_mars_test)
r2_mars_test = r2_score(yTest, pred_mars_test)

# Funciones base seleccionadas (coeficiente != 0)
selected_mask = np.abs(mars_model.coef_) > 1e-8
n_selected = np.sum(selected_mask)

print(f"\nFunciones base seleccionadas: {n_selected} de {B_train.shape[1]}")
print(f"Alpha optimo (LassoCV): {mars_model.alpha_:.6f}")
print(f"\nMSE Test MARS: {mse_mars:.4f}")
print(f"R2 Test MARS:  {r2_mars_test:.4f}")

# --- Mostrar funciones base seleccionadas (cuts y coeficientes) ---
print("\n" + "=" * 70)
print("MARS - Funciones base seleccionadas (cuts)")
print("=" * 70)
print(f"{'#':<4} {'Funcion base':<45} {'Coeficiente':>12}")
print("-" * 70)
print(f"{'0':<4} {'(Intercepto)':<45} {mars_model.intercept_:>12.4f}")
idx_print = 1
for i in range(len(hinge_info)):
    if selected_mask[i]:
        direction, j, t = hinge_info[i]
        name = feature_names[j]
        if direction == 'pos':
            desc = f"h({name} - {t:.4f})"
        else:
            desc = f"h({t:.4f} - {name})"
        print(f"{idx_print:<4} {desc:<45} {mars_model.coef_[i]:>12.4f}")
        idx_print += 1
print("=" * 70)
```

```python
# Importancia de variables MARS
# Se agrupan los coeficientes absolutos por variable
importance = {}
for i in range(len(hinge_info)):
    if selected_mask[i]:
        direction, j, t = hinge_info[i]
        name = feature_names[j]
        importance[name] = importance.get(name, 0) + abs(mars_model.coef_[i])

imp_sorted = dict(sorted(importance.items(), key=lambda x: x[1], reverse=True))

print("Importancia de variables MARS:")
for name, imp in imp_sorted.items():
    print(f"  {name:<25} {imp:.4f}")

# Grafico de barras
plt.figure(figsize=(10, 5))
plt.barh(list(imp_sorted.keys())[::-1],
         list(imp_sorted.values())[::-1],
         color='steelblue', edgecolor='black', linewidth=0.5)
plt.xlabel('Importancia (suma |coeficientes|)', fontsize=12)
plt.title('Importancia de Variables - Modelo MARS',
          fontsize=14, fontweight='bold')
for i, (name, val) in enumerate(list(imp_sorted.items())[::-1]):
    plt.text(val + 0.01, i, f'{val:.3f}', va='center', fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# 3.f) Graficos de Dependencia Parcial MARS: Price (4), Advertising (2), Age (5)

def mars_partial_effect(feature_idx, x_range, hinge_info, coefs):
    """Efecto parcial: suma de hinge functions asociadas a una variable"""
    effect = np.zeros_like(x_range, dtype=float)
    for i, (direction, j, t) in enumerate(hinge_info):
        if j == feature_idx and abs(coefs[i]) > 1e-8:
            if direction == 'pos':
                effect += coefs[i] * np.maximum(0, x_range - t)
            else:
                effect += coefs[i] * np.maximum(0, t - x_range)
    return effect

indices_interes = [4, 2, 5]  # Price, Advertising, Age
nombres_interes = ['Price', 'Advertising', 'Age']

plt.figure(figsize=(18, 5))
for i, (feat_idx, nombre) in enumerate(zip(indices_interes, nombres_interes)):
    plt.subplot(1, 3, i + 1)
    x_min = XTrain_scale[:, feat_idx].min()
    x_max = XTrain_scale[:, feat_idx].max()
    x_vals = np.linspace(x_min, x_max, 300)
    y_effect = mars_partial_effect(feat_idx, x_vals, hinge_info, mars_model.coef_)
    plt.plot(x_vals, y_effect, 'b-', linewidth=2)

    # Marcar knots seleccionados
    knots_sel = set([hinge_info[k][2] for k in range(len(hinge_info))
                     if selected_mask[k] and hinge_info[k][1] == feat_idx])
    for k in knots_sel:
        plt.axvline(x=k, color='red', linestyle='--', alpha=0.5, linewidth=1)

    plt.title(f"Efecto parcial: {nombre} (MARS)",
              fontsize=12, fontweight='bold')
    plt.xlabel(f"{nombre} (estandarizado)")
    plt.ylabel("Contribucion a Sales")
    plt.grid(True, alpha=0.3)

plt.suptitle("Graficos de Dependencia Parcial - Modelo MARS",
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

**3.f) Anlisis e Interpretacin de Predictores MARS**

**Resumen del modelo:**
El modelo MARS (Multivariate Adaptive Regression Splines) construye funciones base tipo bisagra (hinge functions):
- **h(x - t)+ = max(0, x - t)**: Se activa cuando x supera el knot t (tramo derecho)
- **h(t - x)+ = max(0, t - x)**: Se activa cuando x es menor que el knot t (tramo izquierdo)

Esto permite capturar relaciones no lineales mediante segmentos lineales por tramos, conectados en puntos de quiebre (knots). El algoritmo ejecuta un forward pass (agregar funciones de forma greedy) y un backward pass (podar usando GCV para evitar sobreajuste).

**Interpretacin de los 3 predictores seleccionados:**

**Price (Precio del producto):** El grfico de dependencia parcial muestra una relacin negativa con las ventas. Los knots (lneas rojas punteadas) sealan los puntos donde la sensibilidad al precio cambia. Para valores bajos de Price, la contribucin a Sales es positiva (precios bajos impulsan ventas), mientras que para precios altos la contribucin se vuelve fuertemente negativa. Esto es consistente con la ley de la demanda: a mayor precio, menores ventas. Las funciones hinge capturan que la elasticidad no es constante  hay rangos de precio donde los consumidores son ms o menos sensibles al cambio.

**Advertising (Publicidad):** Se observa un efecto positivo que crece con la inversin publicitaria, pero con puntos de quiebre que evidencian rendimientos decrecientes. Los knots delimitan zonas: inicialmente cada unidad adicional de publicidad genera un incremento significativo en ventas, pero pasado cierto umbral, el efecto marginal se reduce (saturacin publicitaria). Este patrn de "rendimientos decrecientes" es una caracterstica clsica que MARS captura naturalmente mediante sus funciones bisagra.

**Age (Edad promedio de la poblacin):** El efecto de la edad muestra una tendencia predominantemente negativa: poblaciones con mayor edad promedio tienden a generar menores ventas de asientos para autos (producto orientado a familias jvenes con nios). Los knots revelan los rangos de edad donde el efecto es ms pronunciado, lo cual tiene sentido demogrfico dado que la demanda de sillas infantiles est concentrada en hogares con nios pequeos.
